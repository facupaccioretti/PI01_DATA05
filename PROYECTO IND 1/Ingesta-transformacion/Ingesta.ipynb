```python
import pandas as pd
import numpy as np
```

### INGESTA DE DATOS


```python
# CREAMOS LOS DATAFRAME E IMPORTAMOS LOS ARCHIVOS 
# CON LOS DATOS QUE LOS CONFORMARÁN:

df_Amazon = pd.read_csv("amazon_prime_titles.csv")
df_Disney = pd.read_csv("disney_plus_titles.csv")
df_Hulu = pd.read_csv("hulu_titles.csv")
df_Netflix = pd.read_json("netflix_titles.json")
```

Utilizamos funciones de pandas para echarle un vistazo a los csv y poder comenzar a realizar una limpieza, transformación y normalización de los datos.

```python
df_Amazon.info() #Nos brinda informacion sobre el dataframe
df_Amazon.isnull().sum() # Nos brinda la cantidad de datos nulos por columna
df_Amazon.head(10) #Nos muestra los 10 primeros registros del dataframe
```

```python
df_Disney.info() #Nos brinda informacion sobre el dataframe
df_Disney.isnull().sum() # Nos brinda la cantidad de datos nulos por columna
df_Disney.head(10) #Nos muestra los 10 primeros registros del dataframe
```

```python
df_Hulu.info() #Nos brinda informacion sobre el dataframe
df_Hulu.isnull().sum() # Nos brinda la cantidad de datos nulos por columna
df_Hulu.head(10) #Nos muestra los 10 primeros registros del dataframe
```

```python
df_Netflix.info() #Nos brinda informacion sobre el dataframe
df_Netflix.isnull().sum() #Nos brinda la cantidad de datos nulos por columna
df_Netflix.head(10) #Nos muestra los 10 primeros registros del dataframe
```

Como podemos observar en los 4 dataframes existen columnas que tienen distinta cantidad de datos con respecto a la columna ID que seria la primary key de nuestras tablas. Además, pudimos observar la cantidad de nulos por cada columna. 



### Tratamiento de valores faltantes

A continuacion rellenaremos todos los valores nulos ingresando la leyenda "sin dato" en su lugar, para asi poder posteriormente trabajar con esto.

```python
df_Amazon.fillna("Sin dato", inplace = True)
df_Disney.fillna("Sin dato", inplace = True)
df_Hulu.fillna("Sin dato", inplace = True)
df_Netflix.fillna("Sin dato", inplace = True)
```

Ahora en lugar de tener valores nulos, tenemos la leyenda "sin dato" donde no existe el dato. Corroboramos esto revisando nuevamente la cantidad de nulos por columna.

```python
print(df_Amazon.isnull().sum())
print(df_Disney.isnull().sum())
print(df_Hulu.isnull().sum())
print(df_Netflix.isnull().sum())
```

A continuacion, revisaremos que no existan ID duplicadas.

```python
df_Amazon["show_id"].duplicated().sum() #Nos brinda informacion acerca de si existen ID duplicadas, esto no deberia suceder.
```

```python
df_Disney["show_id"].duplicated().sum() #Nos brinda informacion acerca de si existen ID duplicadas, esto no deberia suceder.
```

```python
df_Hulu["show_id"].duplicated().sum()#Nos brinda informacion acerca de si existen ID duplicadas, esto no deberia suceder.
```

```python
df_Netflix["show_id"].duplicated().sum()#Nos brinda informacion acerca de si existen ID duplicadas, esto no deberia suceder.
```

Ahora sabemos que no existen claves ID duplicadas, es decir que para cada fila de nuestro dataframe existe un unico id que la representa.

A continuacion, agregaremos la columna "plataforma" en nuestros dataframes, con el fin de indicar a que plataforma pertenece cada registro.

```python
df_Amazon.insert(1,'Plataforma','Amazon Prime')
df_Disney.insert(1,'Plataforma','Disney+')
df_Hulu.insert(1,'Plataforma','Hulu')
df_Netflix.insert(1,'Plataforma','Netflix')
```

Ahora uniremos todos los dataframe en uno solo concatenandolos.

```python
films_df = pd.concat([df_Amazon,df_Disney, df_Hulu, df_Netflix]) #Concatenamos todos los dataframes

```

```python
films_df.drop(['show_id'], axis=1, inplace= True) #Dropeamos la columna showid, no es necesaria ya que pandas nos da el numero de fila
```

```python
films_df
```

Renombramos las columnas:

```python
films_df.columns=["Plataforma", "Tipo", "Titulo", "Director", "Reparto", "Pais", "FechaDeAgregación", "FechaDeEstreno", "Audiencia", "Duracion", "Generos", "Descripcion"]
```

Ahora finalmente exportamos a csv el dataframe final: films_df

```python
ff = films_df['Duracion'].str.split(" ", n = 1, expand = True)
films_df['duracion_film'] = ff[0]
films_df['unidad_duracion'] = ff[1]
films_df
```

```python
#Dropeamos la columna duracion que era la original, porque ya la separamos
films_df.drop(columns='Duracion',inplace = True)
```

```python
#Renombramos la columna
films_df.rename(columns={'duracion_film' : 'Duracion'}, inplace=True)
```

```python
#Reemplazamos ese valor 'Sin' por 1, para que no influya en nuestros resultados
films_df.Duracion.replace({"Sin": 1}, inplace= True)
films_df.Duracion.value_counts()
```

```python
films_df.to_csv('films_df.csv')
```
